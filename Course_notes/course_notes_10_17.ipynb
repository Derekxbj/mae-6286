```python
import matplotlib
import bqplot
import numpy as np
import bqplot.pyplot as bqplt
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 1000)
t = np.linspace(0, 1)

a = 1.0

def q_0(x):
    return np.exp(-200*(x-0.2)**2)

from bqplot import LinearScale, Lines, Axis, Figure
from ipywidgets import IntSlider, FloatSlider, HBox, VBox, Play, jslink

time_interval = 0.01

fig = bqplt.figure()

advect = bqplt.plot(x, q_0(x-a*0))

frame_slider = IntSlider(min=0, max=80, step=1, description='Frame', value=0)

def frame_changed(change):
    t = frame_slider.value*time_interval
    advect.y = q_0(x-a*t)


    
frame_slider.observe(frame_changed, 'value')


play_button = Play(min=0, max=80, step=1)

jslink((play_button, 'value'), (frame_slider, 'value'))
VBox([fig, HBox([frame_slider, play_button])])
```

```python
import time
a = 1.0
m = 50
dx = 1./m
x = np.arange(-dx/2, 1.+dx/2, dx)
t = 0.
T = 0.5
dt = 0.8 * dx / a

Q = np.exp(-200*(x-0.2)**2)
Qnew = np.empty(Q.shape)

plt.plot(x, Q)

while t < T:
    
    Qnew[0] = Q[1]
    Qnew[-1] = Q[-2]
    
    for i in range(1, len(x)):
        Qnew[i] = Q[i] - a*dt/dx * (Q[i]-Q[i-1])
        
    Q = Qnew.copy()
    t = t + dt
    lines[0].y = Q
    plt.plot(x, Q)
    time.sleep(0.1)
```
